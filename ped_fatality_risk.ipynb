```python
import pandas as pd
from risk_calc import * 
```

Example usage

Assumes average height/weight of a 35 y/o male. SUV v Car. 

```python
tefft_profiles = [
    dict(profile_id=3, age_years=35, height_in=66, weight_lb=140, vehicle_is_truck=0),
    dict(profile_id=4, age_years=35, height_in=66, weight_lb=140, vehicle_is_truck=1),
]
mueller_profiles = [
    dict(profile_id=3, age_years=35, vehicle_type="car", sex_male=0, severity="fatal"),
    dict(profile_id=4, age_years=35, vehicle_type="suv", sex_male=0, severity="fatal"),
]

plot_ped_risk(
    model="both",
    tefft_profiles=tefft_profiles,
    mueller_profiles=mueller_profiles,
    units="mph",
    speed_min=5, 
    speed_max=65, 
    speed_step=1,
    pct_markers=(25,50,75),
    match_colors=True,
    match_by="id", 
    id_key="profile_id",
    # include_id_in_label=True,
    show_marker_labels=True,
    marker_label_rotation=30,
    # marker_label_offset=(10, 6),
    out="paired_by_id_with_vtype.jpeg"
)
```
